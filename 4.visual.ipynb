```python
import pandas as pd
from pyecharts import Bar
from pyecharts import Line,  Overlap, Style
```

```python
df = pd.read_csv(r'C:\DATA\KB_capstone\corr_table\corr_table.csv', engine = 'c')
```

```python
df.index = df['variable_name']
```

```python
df = df.iloc[:,2:]
```

```python
df = df.dropna()
df = df.round(2)
```

```python
var = df.T.var()*1000
average = df.T.mean()
```

```python
df['var'] = var
df['mean'] = average
```

```python
df.sort_values(['var'], ascending = False)
```

```python
index = []
for i in df.values:
    i = i[0:-2]
    if max(abs(i)) < 0.2:
        index.append(False)
    else:
        index.append(True)
```

```python
df = df[index]
```

```python
df_var = df.sort_values(['var'],ascending = False)
df_mean =  df.sort_values(['mean'],ascending = False)
```
